```python


rank_lst = ['Rank']*100
name_lst = ['Name']*100
yds_lst = ['Passing_Yds']*100
td_lst = ['TDs']*100
int_lst = ['Int']*100
td_int_lst = ['TD_INT_Ratio']*100
comp_lst = ['Completions']*100
attmpt_lst = ['Attempts']*100
comp_per_lst = ['Completion_Percentage']*100
pass_ptg_lst = ['Passer_Ratings']*100
ypg_lst = ['YPG']*100
gp_lst = ['Game_Played']*100
sp_lst = ['Seasons_Played']*100
num_teams_lst = ['Number_of_Teams']*100



master_rank = list(zip(rank_lst,rank))
master_name = list(zip(name_lst,name))
master_yds = list(zip(yds_lst,yards))
master_td = list(zip(td_lst,tds))
master_int = list(zip(int_lst,ints))
master_ratio = list(zip(td_int_lst,td_int_ratio))
master_comp = list(zip(comp_lst,completions))
master_attmpt = list(zip(attmpt_lst,attempts))
master_comp_per = list(zip(comp_per_lst,comp_per))
master_pass_ptg = list(zip(pass_ptg_lst,passer_rating))
master_ypg = list(zip(ypg_lst,ypg))
master_gp = list(zip(gp_lst,games_played))
master_sp = list(zip(sp_lst,seasons_played))
master_num = list(zip(num_teams_lst,number_teams))

stuff3 = list(zip(*[master_rank,master_name,master_yds,master_td,master_int,master_int,master_ratio,master_comp,master_attmpt,master_comp_per,master_pass_ptg,master_ypg,master_gp,master_sp,master_num]))

lengthyOne = []
for i in range(len(stuff3)):
    lengthyOne.append(dict(stuff3[i]))
    
lengthyOne
```

```python
import pymongo

# Setup connection to mongodb
conn = "mongodb://localhost:27017"
client = pymongo.MongoClient(conn)

# Select database and collection to use
db = client.nfl_qbs
collection = db.football_players

db.collection.insert_many(lengthyOne)
```
